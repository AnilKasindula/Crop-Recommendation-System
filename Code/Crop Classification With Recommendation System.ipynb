
# Scale the features using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()

X_train = ms.fit_transform(X_train)
X_test = ms.transform(X_test)
```

# Standarization

```python
    from sklearn.preprocessing import StandardScaler,MinMaxScaler
    sc = StandardScaler()

    sc.fit(X_train)
    X_train = sc.transform(X_train)
    X_test = sc.transform(X_test)
```

```python
X_train
```

# Training Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score,classification_report

# create instances of all models
models = {
    'Logistic Regression': LogisticRegression(),
    'Naive Bayes': GaussianNB(),
    'Support Vector Machine': SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Bagging': BaggingClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'Extra Trees': ExtraTreeClassifier(),
}

accuracies = []
pred = {}

for name, md in models.items():
    md.fit(X_train,y_train)
    ypred = md.predict(X_test)
    pred[name] = ypred
    accuracies.append(accuracy_score(y_test, ypred))
    print(f"{name}  with accuracy : {accuracy_score(y_test,ypred)}")

for name, ypred in pred.items():
    print("-" * 80)
    print(f"{name} - Accuracy: {accuracy_score(y_test, ypred):.4f}")
    print("Classification Report:")
    print(classification_report(y_test, ypred, zero_division=1))

fig, ax = plt.subplots(figsize=(10, 6))
models_names = list(models.keys())
x = np.arange(len(models_names))
ax.bar(x, accuracies, align='center', alpha=0.7, color='skyblue')
ax.set_xticks(x)
ax.set_xticklabels(models_names, rotation=45, ha='right')
ax.set_ylabel('Accuracy')
ax.set_title('Model Accuracies')
plt.tight_layout()
plt.show()
```

# Predictive System

```python
def recommendation(N,P,k,temperature,humidity,ph,rainfal):
    features = np.array([[N,P,k,temperature,humidity,ph,rainfal]])
    transformed_features = ms.fit_transform(features)
    transformed_features = sc.fit_transform(transformed_features)
    prediction = rfc.predict(transformed_features).reshape(1,-1)
    
    return prediction[0] 
```
